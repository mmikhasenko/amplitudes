# Relativistic Breit-Wigner with dynamical width

The relativistic Breit-Wigner is given by 
$$
BW = \frac{h_l(p)}{m^2-s-i m \Gamma(s)}, \quad \Gamma(s) = \Gamma_0\frac{p m}{p_0\sqrt{s}} \frac{h_1^2(p)}{h_1^2(p_0)}.
$$
The factors $h_l(p)$ include threshold factor $p^l$ and formfactors which control high energy behavior (e.g. Blatt-Weisskopf functions).

For the P-wave, $h_l^2(p) = p^2R^2/(1+p^2R^2)$.

```julia
# Kaellen function
λ(x,y,z) = x^2 + y^2 + z^2 - 2*x*y - 2*y*z - 2*z*x
# break up momentum
p(s,m1sq,m2sq) = sqrt(λ(s,m1sq,m2sq)/(4*s))
# the Breit wigner
function BW(s,msq,G0,m1sq,m2sq,R)
    psq = λ(s,m1sq,m2sq)/(4*s);
    psq0 = λ(msq,m1sq,m2sq)/(4*msq);
    phsp = sqrt(psq/psq0*msq/s);  # normalized phase space
    ff = psq/psq0*(1.+psq0*R^2)/(1.+psq*R^2);  # normalized form factor
    return sqrt(ff)/(msq-s-1im*sqrt(msq)*G0*phsp*ff)
end
```

```julia
using Plots
gr()
```

```julia
# define masses and widthes
mπ=0.14; mp=0.938;
mΔ=1.232; ΓΔ=0.117;
R=5;
# plot
plot(xlabel="Mpip (GeV)", ylabel="amplitude")
plot!(e->real(BW(e^2,mΔ^2,ΓΔ,mπ^2,mp^2,5)), (mπ+mp):0.01:1.6, label="Re")
plot!(e->imag(BW(e^2,mΔ^2,ΓΔ,mπ^2,mp^2,5)), (mπ+mp):0.01:1.6, label="Im")
```

## GlueX kinematics

The reaction $\gamma p \to \pi \Delta_{\to \pi p}$.
The final state is three particles. The phase space depends on 5 variables
 * angles $\Omega_\Delta$ in the CMS frame
 * angles $\Omega_p$ in the $(\pi p)$-frame
 * mass of the $(\pi p)$-system

The differential cross section is given by the equation below, where I drop factors with only $s$-dependence.
$$
% \newcommand{\d}{\mathrm{d}}
\frac{\d \sigma}{\d t \,\d m_{\pi p}^2} \sim \left|\mathcal{M}\right|^2
%\frac{1}{2\lambda^{1/2}(s,0,m_p^2)} 
\frac{\d \phi}{2\pi}\,\d \Phi(\Delta;\pi,p)
$$

I will assume that Jon consider $\pi p$-spectrum, $t$ is fixed, while integrating over other variables.
Then, the line shape of delta is:
$$
\frac{\d \sigma}{\d t \, \d m_{\pi p}} = N
\frac{\lambda^{1/2}(m_{\pi p}^2,m_p^2,m_\pi^2)}{m_{\pi p}} \big|\text{BW}(s)\big|^2,
$$
where $N$ is normalization constant which depends on total energy $\sqrt{s}$, the factor $m_{\pi p}$ from
jacobian kills one in the phase space therefore the break up momentum from the phase space remains.

```julia
dffCS(s,msq,G0,m1sq,m2sq,R)=abs(BW(s,msq,G0,m1sq,m2sq,R))^2*p(s,mπ^2,mp^2)
# plot
plot(xlabel="Mpip (GeV)", ylabel="differential cross section")
plot!(e->dffCS(e^2,mΔ^2,ΓΔ,mπ^2,mp^2,5), linspace((mπ+mp),1.6,100), label="d Sigma / d Mpip")
```
